# Alpha-helix analysis in a trajectory

```python
import morphoscanner as ms
import nglview as nv
import pandas as pd
```

#### Load the data

```python
loader = ms.data_sample.loader.dataloader()
```

### Analysis

```python
traj = ms.trajectory.trajectory(loader.gro_helix, loader.xtc_helix)
```

```python
traj.explore()
```

```python
traj.plot_frame(0)
```

Since we want to analyze each frame of the trajectory, we do not set any `sampling_interval` in the method `compose_database`. 

```python
traj.compose_database()
```

```python
traj.analyze_inLoop()
```

```python
traj.get_data()
```

The database will be empty, since there is a single peptide in this trajectory.

```python
traj.database
```

```python
traj.helix_score()
```

### Visualization

The plot below shows the variation of alpha-helix percentage along the trajectory.

```python
ms.plot.plot.plot_alpha_perc(traj)
```

We can use a backend function to check in which frame our peptide has reached the maximum percentage of alpha-helix conformation.

```python
ms.backend.helix_recognition.get_max_helix_single_peptide(traj, peptide=0)
```

We can plot that frame to visualize the conformation.

```python
%matplotlib notebook
```

```python
traj.plot_frame(112)
```

We can check the `pandas.DataFrame`, output of `helix_score()` method, of that frame. In the MD that we are using there is only one peptide, so we see only one row.

```python
traj.frames[112].results.helix_score
```

It is possible to compose a `pandas.DataFrame` containing the score of a single peptide for each frame of the trajectory.

```python
ms.backend.helix_recognition.build_df_single_pep_helix_score(traj, peptide=0)
```

To visualize the trajectory it is possible to leverage `MDAnalysis` and `nglview` capabilities.

```python
view = nv.show_mdanalysis(traj.universe.select_atoms("protein"))
```
